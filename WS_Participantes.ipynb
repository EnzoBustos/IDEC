Claro que o `seaborn` tem um parâmetro para isso que é o `hue`, esse parâmetro vai colorir seus dados conforme as observações de uma outra variável, observe:

```
sns.scatterplot(x='conta', y='gorjeta', data=gorjeta, hue='fumante')
```

Não preciso nem falar que o `hue` é extremamente personalizável, o parâmetro `hue_order` especifíca a ordem que você quer que a legenda aparece e o parâmetro `palette` pode receber um dicionário que mapeia o rótulo a uma determinada cor específica de sua preferência (Confira nomes das cores [aqui](https://python-graph-gallery.com/100-calling-a-color-with-seaborn/), você também pode usar o código hexadecimal das cores para ter um poder de personalização ainda melhor

Bom, a principal funcionalidade dos gráficos é relacionar `features`, ou seja, comparar atributos dos nossos dados para ver como eles se relacionam entre si e nós já fizemos isso tentando ver se existe uma relação entre peso e altura ou ainda no exemplo entre gorjetas e conta.

Esses gráficos são chamados de gráficos relacionais, e também estão implementados em `seaborn` como `relplot`, a vantagem de se utilizar um `relplot` é que é muito fácil criar `subplots`, que são múltiplos gráficos. Nele podemos especificar o `kind` que é o tipo de gráfico que queremos, como *line*, *scatter*, entre outros.

```
sns.relplot(x='conta', y='gorjeta', data=gorjeta, kind='scatter', col='fumante')
```

Vamos ver agora algumas customizações que você pode fazer no seu `scatterplot`, para deixar ele visualmente mais atrativo e informativo ao usuário, uma primeira modificação é aumentar o tamanho do ponto, para nosso exemplo vamos fazer esse tamanho ser relacionado ao número de pessoas no grupo.

```
sns.relplot(x='conta', y='gorjeta', data=gorjeta, kind='scatter', size='tamanho', hue='tamanho')
```

Outra customização interessante é com o parâmetro `style` que usa um ponto com uma forma diferente para diferenciar os dados.

```
sns.relplot(x='conta', y='gorjeta', data=gorjeta, kind='scatter', style='fumante', hue='fumante')
```

Por último, pode ser interessante mudarmos a transparência do nosso ponto usando o parâmetro `alpha`, isso pode ser útil quando quisermos usar um `scatterplot` e tivermos muitos pontos um por cima do outro, podendo assim diverenciar em que região há mais observações.

```
sns.relplot(x='conta', y='gorjeta', data=gorjeta, kind='scatter', alpha=0.4)
```

Esse gráfico ficou péssimo! Está muito confuso, pois tem muitas variáveis no meu `plot`, então o que fazemos quando nos encontramos com problemas como esse ?

Também é parte do trabalho do ciencista de dados entender um problema, então com um pouco de pesquisa podemos chegar a algumas regiões que são mais importantes para o nosso estudo.

Ficou um pouco melhor mas ainda podemos melhorar, tirando nosso intervalo de confiança (que é essa sombra que está atrás de cada linha) utilizando o parâmetro `ci`, podemos também brincar com as linhas que temos usando os parâmetro `dashes=False`, para usar uma linha sólida em todas e `markers=True`, para ativar marcações em cada ponto e essa marcação é baseada no subgrupo em `style`.

Nota: O intervalo de confiança é mostrado quando temos mais de uma observação por um determinado valor de x e, por padrão ele nos dá 95% para a média (que é a linha em display), podemos modificar usando o parâmetro `ci` e passando diferentes valores como `sd` para o IC de desvio padrão ou ainda com `None` para ele não ser mostrado.

```
sns.relplot(x='Date', y='AveragePrice', data=avocado,
            kind='line', style='region', hue='region',
            ci=None, dashes=False) #teste com os markers também

plt.xticks(rotation=45)
```

Nessa parte vamos passar por cima de `countplots` e `barplots`, eles envolvem geralmente a comparação entre os grupos de alguma variável categórica, para esse módulo usaremos o `catplot()` uma função que funciona mais ou menos como a `relplot()` para dados categóricos, novamente sendo fácil criar `subplots` com os parâmetros `row` e `col`.

Vamos olhar isso mais de perto com o dataset do Titanic, que consta de várias observações sobre os passageiros, como sexo, idade, qual a classe da passagem, onde embarcou entre outros fatores e nossa variável alvo é se eles sobreviveram ou não ao incidente.

```
titanic = pd.read_csv('titanic.csv')
titanic.info()
```

Primeiro, antes de plottarmos nosso `catplot` vamos aplicar um pouquinho de `pandas` para mudarmos os rótulos nos nossos portos, pois eles estão só com as iniciais e isso não é muito legal (ainda mais para a gente que provavelmente não conhece esses portos).

Uma coisa interessante é que você pode mudar a ordem em que as categorias aparecem no seu gráfico com base no parâmetro `order`, que recebe uma lista da ordem que você quer e ainda funcionam outros parâmetro como o `palette` que já tinhamos visto.

```
#Mapeamento das mudanças
portos = {
    'C': 'Cherbourg',
    'Q': 'Queenstown',
    'S': 'Southampton',
}

#Código para mudarmos nossa coluna com base no dicionário
titanic['Embarked'].replace(portos, inplace=True)
```

```
ordem = ['Queenstown', 'Cherbourg', 'Southampton']

cores = {
    'Queenstown': 'black', 
    'Cherbourg': 'gold', 
    'Southampton': 'aquamarine',
}

#Tente usar os parâmetros order e palette para personalizar o seu gráfico
sns.catplot(x='Embarked', data=titanic,
            kind='count')
```

Os `barplot` são um pouco diferentes, eles mostram a média de uma variável quantitativa por categoria. Por exemplo, podemos ver a média dos preços das passagens em 'Fare' baseado no porto em que a pessoa embarcou, pois parece que faz sentido essa relação.

Como vimos também antes, o `seaborn` já automaticamente nos dá um intervalo de confiança de 95% para nossa média e isso pode ser ajustado com o parâmetro `ci`.

```
sns.catplot(x='Embarked', y='Fare', data=titanic, kind='bar')
```

É possível rotacionar seu gráfico trocando os parâmetros x e y, tendo assim um gráfico na horizontal, porém isso não é muito recomendado.

Um gráfico extremamente versátil e útil é o `boxplot` que também pode ser feito com facilidade no `seaborn` e é bem útil por ser uma ferramente estatística forte também na hora de visualizar nossos dados, da mesma forma que o `barplot` ele plotta a relação de uma variável numérica baseado em uma categoria escolhida.

A melhor parte do `boxplot` é que ele nos dá muitas informações, como o quanto seus dados estão espalhados, os quartis e até os `outliers` e facilita muito a comparação entre os grupos.

Nota: você pode omitir os `outliers` com o parâmetro `sym=''` ou ainda alterar o display dele, por padrão os `outliers` são calculados por: $1.5*IQR$

Em que o IQR é a distância interquartis, mas isso pode ser alterado usando o parâmetro `whis` que pode receber um número que multiplicará o IQR ou ainda uma lista que irá considerar os percentis desejados (esse parâmetro controla mais especificamente as linhas cinzas acima e abaixo do `boxplot`)

```
#Veja também a relação entre Survived e Fare
#Utilize o parâmetro sym e whis também
g = sns.catplot(x='Embarked', y='Fare', data=titanic, kind='box')
```

Por último vale mostrar os `pointplots` que são gráficos muito parecidos com os `barplots` só que ao invés de barras são representados apenas pontos e são muito úteis para comparar variáveis categóricas que apresentam alguma ordenação inerente

```
#Vamos agora categorizar nossas idades

divisões = [0, 12, 25, 37, 51, 80]
nomes = ['Criança', 'Jovem', 'Adulto', 'Meia-Idade', 'Idoso']
titanic['Age'] = pd.cut(titanic['Age'], divisões, labels=nomes)
```

```
#Agora vamos fazer uma comparação do barplot com o pointplot

sns.catplot(x='Age', y='Fare', hue='Sex',data=titanic, kind='bar')
sns.catplot(x='Age', y='Fare', hue='Sex',data=titanic, kind='point')
```

Mesmo eles tendo uma ideia muito parecida, o `pointplot` torna muito mais fácil visualizar de viasualizar a inclinações da reta e quando os dados estão um sobre o outro (para de uma ideia do quanto estão acima).

É ainda possível retirar as linhas que ligam os pontos com o parâmetro `join=False` e também pode colocar linhas no final dos intervalos de confiança com o parâmetro `capsize`.

<h1><li> Customizando seus plots </li></h1>

---

<img src='https://drive.google.com/uc?id=1b5FK0muuELXj0NDkvjBVkzD5T_59T874' width=500>

Gráficos de `seaborn` já são bem agradáveis de se olhar, mas por qual motivo então eu deveria personalizar eles ainda mais ?

Bom, isso pode ir desde preferência pessoal até melhorar a legibilidade ou até guiar até sua interpretação.

Existem alguns modelos de estilo já predefinidos que podem ser modificados utilizando o comando `sns.set_style()`, que pode receber algumas opções como: 'white', 'dark', 'whitegrid', 'darkgrid', 'ticks'. Por padrão o `seaborn` utiliza o 'white'.

```
#Mude os estilos aqui
sns.set_style('ticks')
sns.catplot(x='Age', y='Fare', hue='Sex',data=titanic, kind='point')
```

Temos também o comando `sns.set_palette()` que define as cores dos principais elementos, você pode usar uma paleta de cores pré-definida ou ainda criar o seu próprio estilo de cores, através de uma lista de nomes de cores ou de códigos hexadecimais

Você pode conferir algumas das paletas existentes [aqui](https://seaborn.pydata.org/tutorial/color_palettes.html)

Os principais estilos de paletas são:

*   Divergentes: quando os dois extremos tem grande impacto
*   Sequencial: para dados em que um dos extremos importa



```
#Abra a documentação e teste algumas paletas
sns.set_palette('mako')
sns.catplot(x='Age', y='Fare',data=titanic, kind='bar', ci=None)
```

Também podemos mudar a escala do nosso plot com a função `sns.set_context()` que estaria especificando o contexto de apresentação desse gráfico, do menor para o maior podemos colocar os parâmetros: 'paper', 'notebook', 'talk', 'poster'.

Por padrão o `seaborn` utiliza 'paper'

```
sns.set_context('paper')
sns.relplot(x='conta', y='gorjeta', data=gorjeta, kind='scatter', size='tamanho', hue='tamanho')
```

Uma das informações mais cruciais na hora de passar a informação sobre o seu gráfico é definir um bom título e também subtítulos para cada um dos eixos, de forma que o gráfico por si só passe a sua informação e quem estiver vendo o gráfico possa receber a mensagem claramente.

Não querendo entrar muito em detalhes, mas o `seaborn` devolve dois tipos de objetos de Python, os `FacetGrid` e `AxesSubplot`, cada um tem suas peculiaridades e o importante é que `catplot` e `relplot` devolvem `FacetGrid`, que suportam múltiplos gráficos em uma única imagem, enquanto o `AxesSubplot` suporta apenas 1 gráfico.

Como vamos trabalhar com `FacetGrid` é mais fácil usar uma variável para fazer as mudanças de título e legendas

```
sns.set_palette('bright')

g = sns.catplot(x='Age', y='Fare', data=titanic, kind='box', sym='', col='Survived')

#Ajusta o título da figura como um todo e y controla a altura desse título central
g.fig.suptitle('Comparação entre Sobreviventes do Titanic', y=1.05)


#Infelizmente não achei um modo direto de alterar os títulos dos subgráficos a não ser assim :/
axes = g.axes.flatten() #aqui você está quebrando seu gráfico em cada um dos seus subplots
axes[0].set_title("Não Sobreviveu")
axes[1].set_title("Sobreviveu")

'''
#Ou ainda alterar esses valores no seu próprio DataFrame já que é uma variável categórica, antes de plottar
mapa = {0: 'Não Sobreviveu',
        1: 'Sobreviveu',
}

titanic['Survived'].replace(mapa, inplace=True)
'''
#Ajusta as legendas em X e em Y
g.set(xlabel='Idade',
      ylabel='Tarifa')

#Para rotacionar as legendas
g.set_xticklabels(rotation=30)

plt.show()
```

Vale aqui uma observação para o caso do seu gráfico ser um `AxesSubplot`, para alterar o título da imagem você usa `g.set_title()` que recebe uma string e também pode receber um parâmetro `y`; para alterar as rotações das legendas é possível usar `plt.xticks(rotation=)`
